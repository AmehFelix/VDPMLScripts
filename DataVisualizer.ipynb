```python
df_Y = pd.DataFrame(columns = ['Fast', 'Above_Avg', 'Avg', 'Below_Avg', 'Slow'])

for tag in Y:
  if tag == 'Fast':
    df_Y = df_Y.append({'Fast' : 1, 'Above_Avg' : 0, 'Avg' : 0, 'Below_Avg' : 0, 'Slow' : 0},
        ignore_index = True)
  elif tag == 'Above_Avg':
    df_Y = df_Y.append({'Fast' : 0, 'Above_Avg' : 1, 'Avg' : 0, 'Below_Avg' : 0, 'Slow' : 0},
        ignore_index = True)
  elif tag == 'Avg':
    df_Y = df_Y.append({'Fast' : 0, 'Above_Avg' : 0, 'Avg' : 1, 'Below_Avg' : 0, 'Slow' : 0},
        ignore_index = True)
  elif tag == 'Below_Avg':
    df_Y = df_Y.append({'Fast' : 0, 'Above_Avg' : 0, 'Avg' : 0, 'Below_Avg' : 1, 'Slow' : 0},
        ignore_index = True)
  else:
    df_Y = df_Y.append({'Fast' : 0, 'Above_Avg' : 0, 'Avg' : 0, 'Below_Avg' : 0, 'Slow' : 1},
        ignore_index = True)
```
